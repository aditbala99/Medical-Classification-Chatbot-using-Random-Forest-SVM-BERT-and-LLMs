```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import string
from datasets import load_from_disk

# Download necessary NLTK resources
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')

# Load the datasets from the specified paths
train_dataset = load_from_disk("/content/medmcqa_dataset/train/")
test_dataset = load_from_disk("/content/medmcqa_dataset/test/")
validation_dataset = load_from_disk("/content/medmcqa_dataset/validation/")

# Define preprocessing function
def preprocess_text(text):
    if not isinstance(text, str):
        return ""  # or some default value based on your use case

    # Tokenization
    tokens = nltk.word_tokenize(text)

    # Lowercasing
    tokens = [token.lower() for token in tokens]

    # Removing Stop Words
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]

    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens]

    # Removing Special Characters and Numbers
    tokens = [token for token in tokens if token.isalpha()]

    return " ".join(tokens)

# Apply preprocessing to each dataset
for dataset, split in zip([train_dataset, test_dataset, validation_dataset], ["train", "test", "validation"]):
    preprocessed_dataset = dataset.map(lambda example: {
        'question': preprocess_text(example['question']),
        'opa': preprocess_text(example['opa']),
        'opb': preprocess_text(example['opb']),
        'opc': preprocess_text(example['opc']),
        'opd': preprocess_text(example['opd']),
        'cop': example['cop'],
        'choice_type': example['choice_type'],
        'exp': preprocess_text(example['exp']),
        'subject_name': example['subject_name'],
        'topic_name': example['topic_name'],
    })

    # Save the preprocessed dataset to a different path
    preprocessed_dataset.save_to_disk(f"/content/medmcqa_preprocessed_dataset/{split}/")

```
